```python
import scanpy as sc
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
import bbknn
import harmonypy as hm
import scanpy.external as sce
```

```python
# Загрузка данных
adata = sc.read_10x_h5("V1_Adult_Mouse_Brain_filtered_feature_bc_matrix.h5")
adata.var_names_make_unique()
```

```python
#функция для сэмплирования UMI
def sample_umi(adata, sampling_rate):
    sampled_data = []
    for i in range(adata.X.shape[0]):
        cell_umi = adata.X[i, :].toarray()[0]
        total_umi_count = np.sum(cell_umi)
        num_sampled_umi = int(total_umi_count * sampling_rate)

        if num_sampled_umi == 0:
            sampled_cell_umi = np.zeros_like(cell_umi)
        else:
            # Создаем список всех UMI (индексы) для конкретной клетки.
            all_umi_indices = []
            for gene_index, umi_count in enumerate(cell_umi):
                all_umi_indices.extend([gene_index] * int(umi_count))

            # Выбираем случайные индексы UMI
            sampled_umi_indices = np.random.choice(all_umi_indices, size=num_sampled_umi, replace=False)

            # подсчитываем выбранные umi
            sampled_cell_umi = np.zeros_like(cell_umi)
            for gene_index in sampled_umi_indices:
                sampled_cell_umi[gene_index]+=1

        sampled_data.append(sampled_cell_umi)
    return np.array(sampled_data)
```

```python
# Уровни сэмплирования
sampling_rates = [0.1, 0.25]
sampled_adatas = [adata.copy()]
sampled_names = ['original']


# Сэмплирование данных и сохранение их в виде разных объектов adata
for rate in sampling_rates:
    print(f"Сэмплирование UMI {rate * 100:.0f}%")
    sampled_adata = adata.copy()
    total_umi_before = sampled_adata.X.sum(axis=1)
    sampled_adata.X = sample_umi(sampled_adata,rate)
    total_umi_after = sampled_adata.X.sum(axis=1)
    sampled_adata.obs['batch'] = f"{rate * 100:.0f}%"
    sampled_adatas.append(sampled_adata)
    sampled_names.append(f"{rate * 100:.0f}%")



    print("Подтверждение уменьшения UMI для первых 3-х клеток...")
    for i in range(min(3, sampled_adata.X.shape[0])): # Ограничение до 3 клеток
      if total_umi_before[i] != 0:
          reduction =  float(total_umi_before[i] / total_umi_after[i])
          print(f"Клетка {i}: UMI до сэмплирования: {total_umi_before[i]}, UMI после сэмплирования: {total_umi_after[i]}, уменьшение в {reduction:.2f} раз.")
      else:
          print(f"Клетка {i}: UMI до сэмплирования: {total_umi_before[i]}, UMI после сэмплирования: {total_umi_after[i]}.")
```

```python
# Визуализация распределения UMI
plt.figure(figsize=(12, 8))
sns.histplot(adata.X.sum(axis=1).A.flatten(), kde=True, label='original', alpha=0.4) # Оригинальное распределение
for sampled_name, sampled_adata in zip(sampled_names[1:], sampled_adatas[1:]):
    sns.histplot(sampled_adata.X.sum(axis=1).flatten(), kde=True, label=sampled_name, alpha=0.4) # Сабсэмплированные
plt.xlabel('Количество UMI')
plt.ylabel('Частота')
plt.title('Распределение UMI в оригинальных и сабсэмплированных данных')
plt.legend()
plt.show()
```

исходя из этого графика, можно сделать вывод, что UMI набирались случайно и равномерно.
 Сэмплирование не искажает форму распределения UMI, а просто смещает его в сторону меньших значений. Это говорит о том, что UMI отбирались с одинаковой вероятностью, независимо от гена. Если бы UMI отбирались не случайно, а с весами, зависящими от генов, то мы увидели бы искажения формы распределения.
То, что общая форма распределения остается похожей для всех уровней сэмплирования, указывает на то, что процедура сэмплирования была честной, и мы действительно моделируем разную глубину секвенирования.
 Видно, что при уменьшении уровня сэмплирования, количество UMI уменьшается, но при этом распределение смещается влево, без изменения общей формы.

```python
# Объединение всех объектов adata
adata_merged1 = sampled_adatas[0].concatenate(*sampled_adatas[1:], batch_categories=sampled_names)


# Нормализация, масштабирование, PCA и UMAP
sc.pp.normalize_total(adata_merged1, target_sum=1e4)
sc.pp.log1p(adata_merged1)
sc.pp.scale(adata_merged1)
sc.tl.pca(adata_merged1, svd_solver='arpack')

#  UMAP до коррекции батч-эффекта
sc.pp.neighbors(adata_merged1, n_neighbors=10, n_pcs=40)
sc.tl.umap(adata_merged1)
sc.pl.umap(adata_merged1, color='batch', title='UMAP до коррекции батч-эффекта')


sc.tl.leiden(adata_merged1, resolution=0.5)
```

```python
# 5. Дифференциальная экспрессия внутри одного типа клеток между батчами
# Выберем один кластер для анализа
selected_cluster = '0'

# Выделение клеток выбранного кластера
cluster_adata = adata_merged1[adata_merged1.obs['leiden'] == selected_cluster]

print(f"Дифференциальная экспрессия генов в кластере {selected_cluster} между батчами...")

# Выполнение дифференциальной экспрессии
sc.tl.rank_genes_groups(cluster_adata, groupby='batch', method='wilcoxon')
sc.pl.rank_genes_groups(cluster_adata, n_genes=20, sharey=False, key = 'rank_genes_groups')
plt.show()
```

Эти графики представляют собой результаты дифференциальной экспрессии генов. Каждый график показывает список генов, которые дифференциально экспрессируются в одном батче по сравнению со всеми остальными.

```python
# Сабсэмплы (10% и 25%)
adata_10 = adata.copy()
adata_10.X = sample_umi(adata, 0.1)
adata_10.obs['batch'] = '10%'

adata_25 = adata.copy()
adata_25.X = sample_umi(adata, 0.25)
adata_25.obs['batch'] = '25%'
```

```python
def process_and_diff_exp(adata, subset_name):

    sc.pp.filter_genes(adata, min_counts=1)
    sc.pp.normalize_total(adata, target_sum=1e4)
    sc.pp.log1p(adata)
    sc.pp.highly_variable_genes(adata, flavor="cell_ranger", n_top_genes=2000)
    adata = adata[:, adata.var.highly_variable]
    sc.tl.pca(adata, n_comps=30)
    sc.pp.neighbors(adata, n_pcs=30) # !!! Добавлено для каждого сабсэмпла !!!
    sc.tl.umap(adata)

    sc.tl.leiden(adata, resolution=0.5)

    sc.tl.rank_genes_groups(adata, groupby="leiden", method='t-test_overestim_var', groups=['0', '1'], key_added=f'rank_genes_groups_{subset_name}')

    # Вывод результатов для кластеров 0 и 1
    print(f"Топ 10 генов для кластера 0 (vs. кластер 1) в {subset_name}:")
    res_df = sc.get.rank_genes_groups_df(adata, group='0', key=f'rank_genes_groups_{subset_name}')
    print(res_df.head(10))

    print(f"Топ 10 генов для кластера 1 (vs. кластер 0) в {subset_name}:")
    res_df = sc.get.rank_genes_groups_df(adata, group='1', key=f'rank_genes_groups_{subset_name}')
    print(res_df.head(10))


    # Вулканический график для кластера 0
    res_df_0 = sc.get.rank_genes_groups_df(adata, group='0', key=f'rank_genes_groups_{subset_name}')
    plt.figure(figsize=(10, 5))
    plt.scatter(res_df_0['logfoldchanges'], -np.log10(res_df_0['pvals']), s=5, alpha=0.5)
    plt.xlabel("log2(fold change)")
    plt.ylabel("-log10(pval)")
    plt.title(f"Вулканический график для кластера 0 в {subset_name}")
    plt.show()


    # Вулканический график для кластера 1
    res_df_1 = sc.get.rank_genes_groups_df(adata, group='1', key=f'rank_genes_groups_{subset_name}')
    plt.figure(figsize=(10, 5))
    plt.scatter(res_df_1['logfoldchanges'], -np.log10(res_df_1['pvals']), s=5, alpha=0.5)
    plt.xlabel("log2(fold change)")
    plt.ylabel("-log10(pval)")
    plt.title(f"Вулканический график для кластера 1 в {subset_name}")
    plt.show()

    return adata

# Обработка и анализ дифференциальной экспрессии для 10% и 25% подмножества
adata_10 = process_and_diff_exp(adata_10, '10%')
adata_25 = process_and_diff_exp(adata_25, '25%')




# Объединение
adata_merged = adata_10.concatenate(adata_25)
sc.pp.neighbors(adata_merged, n_pcs = 30)
sc.tl.leiden(adata_merged, resolution=0.5)
```

```python
# Визуализация
sc.pl.umap(adata_merged, color=['batch', 'leiden'], title='Объединенный UMAP по батчам')
plt.show()

# Визуализация отдельных UMAP
sc.pl.umap(adata_10, color=['batch', 'leiden'], title='UMAP для 10% подмножества')
plt.show()
sc.pl.umap(adata_25, color=['batch', 'leiden'], title='UMAP для 25% подмножества')
plt.show()
```

```python
# Stacked bar plot до коррекции (Harmony)
df_before = adata_merged.obs.groupby(["leiden", "batch"]).size().unstack().T
fig, ax = plt.subplots(figsize=(7, 3))
df_before.plot(kind="bar", stacked=True, width=1, edgecolor="black", ax=ax)
ax.legend(loc=(1.01, 0.5))
plt.title('Stacked Bar plot до коррекции (Harmony)')
plt.show()
```

```python
# Коррекция батч-эффекта с помощью bbknn
bbknn.bbknn(adata_merged, batch_key="batch")
print("bbknn завершен.")
sc.tl.umap(adata_merged)
sc.pl.umap(adata_merged, color="batch", title='UMAP после коррекции батч-эффекта (bbknn)')
plt.show()


# Коррекция батч-эффекта с помощью Harmony
sce.pp.harmony_integrate(adata_merged, key="batch", max_iter_harmony=30)
sc.pp.neighbors(adata_merged, use_rep="X_pca_harmony")
sc.tl.umap(adata_merged)

sc.pl.umap(adata_merged, color="batch", title='UMAP после коррекции батч-эффекта (Harmony)')
```

```python
# Кластеризация для объединенного объекта (после коррекции bbknn)
sc.tl.leiden(adata_merged, resolution=0.5)
sc.pl.umap(adata_merged, color='leiden', title='UMAP (bbknn) c кластерами')
plt.show()

# Кластеризация для объединенного объекта (после коррекции Harmony)
adata_harmony = adata_merged.copy()
sc.tl.leiden(adata_harmony, resolution=0.5)
sc.pl.umap(adata_harmony, color='leiden', title='UMAP (Harmony) c кластерами')
plt.show()
```

```python
# Stacked bar plot (после Harmony)
df_harmony = adata_merged.obs.groupby(["leiden", "batch"]).size().unstack().T
fig, ax = plt.subplots(figsize=(7, 3))
df_harmony.plot(kind="bar", stacked=True, width=1, edgecolor="black", ax=ax)
ax.legend(loc=(1.01, 0.5))
plt.title('Stacked Bar plot после коррекции (Harmony)')
plt.show()
```

UMAP до коррекции показывает, что сэмплы с разной глубиной секвенирования образуют отдельные области.
UMAP после коррекции показывает, что эти области перемешиваются.
Кластеризация после коррекции показывает, что клетки с разной глубиной секвенирования кластеризуются вместе.
Stacked Bar plot показывает, что “батчи” равномерно распределены по кластерам после коррекции.
